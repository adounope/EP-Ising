```python
import numpy as np
import matplotlib.pyplot as plt
import os
%load_ext autoreload
%autoreload 2
#%matplotlib inline
%matplotlib ipympl
```

```python
folder_name = './results/104_2025-05-05T23:06_MNIST_1000_step_test_b8'
step_list = [2,3,4,8,16, 32, 64, 128, 256, 512, 1024]


acc_historys = np.empty((len(step_list), 100, 2))
for i in range(len(step_list)):
    acc_historys[i] = np.load(f'{folder_name}/{step_list[i]}/acc_history.npy')
```

```python
log_N_anneal_step_list = []
epoch_list = []
acc_list = []

num_color = 100
color_list = []
colors = plt.cm.jet(np.linspace(0,1,num_color))

for idx_epoch in range(100):
    for step_idx in range(len(step_list)):
        acc_list.append(acc_historys[step_idx, idx_epoch, 1])
        epoch_list.append(idx_epoch)

        log_N_anneal_step_list.append(np.log2(step_list[step_idx]))
        color_list.append(colors[int(acc_historys[step_idx, idx_epoch, 0]*100)%num_color])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(log_N_anneal_step_list, epoch_list, acc_list, color = color_list)
ax.set_xlabel('$log_2$(anneal step)', fontsize=13)
ax.set_ylabel('epoch', fontsize=13)
ax.set_zlabel('validation acc', fontsize=13)
ax.set_xticks(np.arange(1, 11))
plt.show()
```
